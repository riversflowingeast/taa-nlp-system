```python
import fitz
import io
from IPython.display import display
from PIL import Image
import pytesseract
import pandas as pd
```

Loading first page of PDF as an image and test extraction using Tesseract's optical character recognition.

```python
doc = fitz.open("../data/external/Xoo-Dictionary.pdf")
page = doc[16]

# Use a matrix to render PDF page at higher resolution
zoom = 3  # 3 = 300%, for target 300 DPI
mat = fitz.Matrix(zoom, zoom)
pix = page.get_pixmap(matrix=mat)

img = Image.open(io.BytesIO(pix.tobytes("png")))
print("Rendered image size:", img.size)
display(img)
```

```python
print(img.size)
```

**Note:**  
- Tesseract does not have a native OCR model for Khoisan languages.  
- Partial recognition may be possible using Xhosa (`xho`) or Zulu (`zul`) language packs, as they include some click consonants.  
- For more accurate results on Khoisan text, a **custom Tesseract model** trained on the specific characters may be needed.

```python
# Extract text from image
text = pytesseract.image_to_string(img, lang="eng")
print(text)
```
